```
import numpy as np
import pandas as pd
import nlplot
import plotly
from plotly.subplots import make_subplots

pd.set_option('display.max_columns', 300)
pd.set_option('display.max_rows', 300)
pd.options.display.float_format = '{:.3f}'.format
pd.set_option('display.max_colwidth', 5000)
```

```
df = pd.read_csv('kyufu.csv')
```

```
display(df.head(3), df.tail(3), df.shape)
```

```
npt = nlplot.NLPlot(df, taget_col='text')
```

```
stopwords = npt.get_stopword(top_n=2, min_freq=0)
```

```
npt.bar_ngram(
    title='持続化給付金',
    xaxis_label='word_count',
    yaxis_label='word',
    ngram=1,
    top_n=50,
    stopwords=stopwords,
)
```

```
npt.wordcloud(
    max_words=100,
    max_font_size=100,
    colormap='tab20_r',
    stopwords=stopwords,
)
```

```
# ビルド
npt.build_graph(stopwords=stopwords, min_edge_frequency=25)

# ビルド後にノードとエッジの数が表示される。この数が100前後になるようにすると、ネットワークが綺麗に描画できます
node_size:63
edge_size:63

npt.co_network(
    title='持続化給付金',
    sizing=100,
    node_size='adjacency_frequency',
    color_palette='hls',
    width=1600,
    height=1100,
    save=False
)
```
